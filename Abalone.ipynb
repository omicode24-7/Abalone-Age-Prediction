```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC,SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
X=df_new.drop('Rings',axis=1)
Y=df_new['Rings']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
# we have very less data for few categories therefore we will use sampling technique over the dataset
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros=RandomOverSampler()
```

```python
X_train1,Y_train1=ros.fit_resample(X_train,Y_train)
X_test1,Y_test1=ros.fit_resample(X_test,Y_test)
```

```python
def create_model(model):
    model.fit(X_train1,Y_train1)
    Y_pred=model.predict(X_test1)
    print("Mean squared error: ",mean_squared_error(Y_test1,Y_pred))
    print("r2 score",r2_score(Y_test1,Y_pred))
```

```python
lr=LinearRegression()
model=create_model(lr)
```

```python
dt1=DecisionTreeClassifier()
model=create_model(dt1)
```

```python
dt1=DecisionTreeClassifier(max_depth=6)
model=create_model(dt1)
```

```python
dt1=DecisionTreeClassifier(min_samples_leaf=35)
model=create_model(dt1)
```

```python
dt2=DecisionTreeClassifier(criterion='entropy')
model=create_model(dt2)
```

```python
dt2=DecisionTreeClassifier(criterion='entropy',max_depth=6)
model=create_model(dt2)
```

```python
dt2=DecisionTreeClassifier(criterion='entropy',min_samples_leaf=35)
model=create_model(dt2)
```

```python
rbf=RandomForestClassifier()
model=create_model(rbf)
```

```python
svc=LinearSVC(random_state=1)
model=create_model(svc)
```

```python
svc=LinearSVC(random_state=1,C=0.05)
model=create_model(svc)
```

```python
psvc=SVC(kernel='poly',random_state=1)
model=create_model(psvc)
```

```python
rsvc=SVC(kernel='rbf',random_state=1)
model=create_model(rsvc)
```

```python
columns=[]
for col in X:
    columns.append(col)
    X_new=df_new[columns]
    X_train,X_test,Y_train,Y_test=train_test_split(X_new,Y,test_size=0.3,random_state=1)
    X_train1,Y_train1=ros.fit_resample(X_train,Y_train)
    X_test1,Y_test1=ros.fit_resample(X_test,Y_test)
    lr=LinearRegression()
    lr.fit(X_train1,Y_train1)
    score=lr.score(X_test1,Y_test1)
    print(columns,":",np.round(score,3))
```
